##### Here we use normalization technique for convert them all into range 0 to 1 for optimization

```python
# x=np.array([df["year"],df["km"],df["color"]])

temp = df["year"].astype("float")
temp2 = df["color"].astype("float")

x = np.array((temp, df["km"], temp2))
print(x.shape)
# Min-Max Normalization per row (feature)
print('x', x[::, 0])
x_min = x.min(axis=1, keepdims=True)
print('x_min', x_min)
x_max = x.max(axis=1, keepdims=True)
print('x_max', x_max)
x_scaled = (x - x_min) / (x_max - x_min)

# Normalize your target array as well to prevent huge gradient steps
y = np.array(df["price"], dtype=float)
y_min = y.min()
y_max = y.max()
y_scaled = (y - y_min) / (y_max - y_min)

print(x_scaled.shape)
print(y_scaled.shape)
```

```python
y=np.array(df["price"])
y
```

##### Full is used to fill values based on its length

```python
w=np.full(len(x), 0.5)
b=0.5
```

##### Perdict Using Dot product 

```python
def predict(x,y,w,b):
    return np.dot(w,x) +b
```

```python
y_pred=predict(x,y,w,b)
y_pred
```

##### Error

```python
def error(y,y_pred):
    return y-y_pred
```

```python
err=error(y,y_pred)
err
```

##### Loss 

```python
def MSE(err):
    return np.mean((err**2))
```

```python
mse=MSE(err)
print(mse)
```

#### Gradient Descent Optimization

```python

def grad(x,y,w,b):
    n=x.shape[1]
    y_pred=predict(x,y,w,b)
    err=error(y,y_pred)
   
    dw = (-2/n) * np.dot(x, (err))
    db = (-2/n) * (np.sum(err))


    return dw,db


```

##### Update Weight and bias

```python
lr=0.01   #Learning rate 
def update(w,b,dw,db,lr):
    w=w- lr*dw
    b=b- lr*db
    return w,b
```

```python
epochs= 3000
for i in range(epochs):
    
    y_pred = predict(x_scaled, y_scaled, w, b)
    
    err = error(y_scaled, y_pred)
    
    mse = MSE(err)
    
    dw, db = grad(x_scaled, y_scaled, w, b)
    

    w, b = update(w, b, dw, db, lr)
    
    if i % 100 == 0:
        print(f"Epoch {i} -> Loss (MSE): {mse:.6f} | Weights: {w} | Bias: {b}")
# print(y)
# print(y_pred)
```

```python
# 1. Raw Test Sample
test_raw = np.array([[2025.0], [69000.0], [0.0]])

# 2. Scale using original training parameters (x_min, x_max)
test_scaled = (test_raw - x_min) / (x_max - x_min)
test_scaled = test_scaled.flatten() # Flatten to 1D array of shape (3,)

# 3. Predict scaled target price
scaled_pred = predict(test_scaled, None, w, b)

# 4. Invert scaling to read the real price in currency
final_price = (scaled_pred * (y_max - y_min)) + y_min
print(f"\nPredicted Car Price: {final_price:.2f}")
print(f"Actual Car Price: {y[0]}")
```
