### approximate weight and bias choose from minimum

````python

w=0.5
b=0.5
````

##### prediction = w*x+b

````python

def predict(x,y,w,b):
    return w*x+b
````

````python
y_pred=predict(x,y,w,b)
y_pred
````

### ERROR = ACTUAL - PREDICT

````python

def error(y,y_pred):
    return y-y_pred
````

````python
err=error(y,y_pred)
err
````

#### LOSS FUNCTION 

##### ```(MEAN SQUARE ERROR) MSE = (1/n) * Σ(y - ŷ)²```

````python
def MSE(err):
    return np.mean((err**2))
````

````python
mse=MSE(err)
print(mse)
````

#### OPTIMIZATION GRADIENT DESCENT

##### DERIVATIONS USING CHAIN RULE

````python
n=len(x)
def grad(x,y,w,b):
    y_pred=predict(x,y,w,b)
    err=error(y,y_pred)
   
    
    dw = -2/n * (np.sum((err) * (x)))
    db = -2/n * (np.sum(err))


    return dw,db


````

##### UPDATE WEIGHT AND BIAS

````python
lr=0.001
def update(w,b,dw,db,lr):
    w=w- lr*dw
    b=b- lr*db
    return w,b
````

##### ITERATION TO FIND BEST WEIGHT AND BIAS MINIMIZE ERROR TO FIT BEST STRAIGHT LINE

````python
epochs=1000
for i in range(epochs):
    y_pred=predict(x,y,w,b)
    err=error(y,y_pred)
    mse=MSE(err)
    dw,db=grad(x,y,w,b)
    w,b=update(w,b,dw,db,lr)

    print(f"weight{w},bias{b}")
print(y)
print(y_pred)
````
